# Atividade 1 - Fitbit Dataset
---
Objetivo: aplicar pipeline de Machine Learning até gerar `Y_pred` com base no dataset Fitbit.

```python
# 1. Importar bibliotecas
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from pathlib import Path
```

```python
# 2. Carregar o dataset
# Troque pelo nome correto do arquivo CSV que você baixou do Kaggle
path = Path("mturkfitbit_export_3.12.16-4.11.16") / \
       "Fitabase Data 3.12.16-4.11.16" / \
       "dailyActivity_merged.csv"

df = pd.read_csv(path)
```

```python
# 3. Explorar os dados
print(df.head())
print(df.info())
print(df.describe())
```

```python
# 4. Definir a variável alvo (Y)
# Critério: usuários com mais de 7500 passos/dia = ativos
df["ActiveUser"] = np.where(df["TotalSteps"] >= 7500, 1, 0)

# Y (alvo)
Y = df["ActiveUser"]
```

```python
# 5. Definir as features (X)
X = df[["TotalSteps", "TotalDistance", "Calories", 
        "VeryActiveMinutes", "FairlyActiveMinutes", "LightlyActiveMinutes", "SedentaryMinutes"]]
```

```python
# 6. Dividir em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# 7. Treinar o modelo
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
```

```python
# 8. Gerar previsões (Y_pred)
Y_pred = model.predict(X_test)
print("Previsões:", Y_pred[:10])
```

```python
# 9. Avaliar o modelo (relatório de classificação)
print(classification_report(y_test, Y_pred))
```
